### Thickness vs Volume
- As the overall size of the particle increases (as measured by volume), the stone particles stay less thick.
- The stone particles appear to be more responsible for large outliers in volume.
- In other words, the stone particles are large and thin, like shards.

```python
# joint plot of thickness and volume
sns.jointplot(
    data = train_data
    ,y = 'fthickness'
    ,x = 'volume'
    ,hue = 'stone_soil'
    ,alpha = 0.5
)
```

### Thickness vs Perimeter
- Another expression of the large thin aspect of the stone particles

```python
# joint plot of thickness and volume
sns.jointplot(
    data = train_data
    ,y = 'fthickness'
    ,x = 'perimeter'
    ,hue = 'stone_soil'
    ,alpha = 0.5
)
```

### Thickness vs Transparency
- It's largely mixed for the less thick particles, but at higher thicknesses the stone particles are noticeably more transparent.

```python
# joint plot of thickness and volume
sns.jointplot(
    data = train_data
    ,x = 'fthickness'
    ,y = 'transparency'
    ,hue = 'stone_soil'
    ,alpha = 0.5
)
```

### Roundness vs Ellipticity
- I'm not sure what this relationship indicates, but there's a clear grouping of stone particles.

```python
# joint plot of roundness and ellipticity
sns.jointplot(
    data = train_data
    ,y = 'roundness'
    ,x = 'ellipticity'
    ,hue = 'stone_soil'
    ,alpha = 0.5
)
```

## Size Relationships
- There are lots of size measurements, so I plotted all of them.

```python
# size measures
size_vars = ['flength'
             ,'fwidth'
             ,'fthickness'
             ,'area'
             ,'perimeter'
             ,'volume']

# pairwise size relationships
sns.pairplot(data = train_data, hue = 'stone_soil', vars = size_vars, corner = True, plot_kws={'alpha':0.5})
```
