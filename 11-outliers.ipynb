```python
# packages
import numpy as np
import pandas as pd
import janitor
import matplotlib.pyplot as plt
```

### TODO: Define outliers of concern

```python
# read data files and clean names

# archaeological soil samples (i.e. dirt)
soil_data = pd.read_csv('data/data_raw/archaeological_soil_data.csv', sep = ',', skiprows=[1]).clean_names()
print('soil',soil_data.shape)

# lithic experimental samples (i.e. stone particles)
stone_data = pd.read_csv('data/data_raw/lithic_experimental_data.csv', sep = ',', skiprows=[1]).clean_names()
print('stone',stone_data.shape)
```

### Soil Data

```python
# numeric variables
soil_data_numerical = soil_data.select_dtypes(include = [np.number])
soil_data_numerical.head(5).T
```

```python
# soil stats and boxplots
for i in range(2,soil_data_numerical.shape[1]):
    print(soil_data_numerical.iloc[:,i].describe())
    soil_data_numerical.iloc[:,i].plot.box()
    plt.show()
```

### Stone Data

```python
# numeric variables
stone_data_numerical = stone_data.select_dtypes(include = [np.number])
stone_data_numerical.head(5).T
```

```python
# stone stats and boxplots
for i in range(2,stone_data_numerical.shape[1]):
    print(stone_data_numerical.iloc[:,i].describe())
    stone_data_numerical.iloc[:,i].plot.box()
    plt.show()
```
