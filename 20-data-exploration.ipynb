```python
# load modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# print column names of stone and soil
print(stone_data.columns)
print(soil_data.columns)
```

```python
# stone as red, soil as blue, plot each column
count = 0
plt.figure(figsize=(30, 30), dpi=100)
colplots = []
for col in stone_data.columns:
    colplots.append(plt.subplot2grid((6,8),(count//8,count%8),colspan=1,rowspan=1))
    colplots[count].set_title("["+str(count)+"] " + col)
    colplots[count].hist(stone_data[col], density=True, color='red', alpha=0.5)
    colplots[count].hist(soil_data[col], density=True, color='blue', alpha=0.5)
    count+=1
```

```python
# using combination of selected columns
select_cols = ['da','dp','sphericity','t_l_aspect_ratio','compactness','solidity','concavity','transparency','curvature','sieve','angularity','ellipticity']
count=0
plt.figure(figsize=(50, 50), dpi=100)
subplots = []
for x_col in range(len(select_cols)):
    for y_col in range(x_col+1,len(select_cols)):
        subplots.append(plt.subplot2grid((12,12),(x_col,y_col),colspan=1,rowspan=1))
        subplots[count].set_title("["+str(count)+"] " + select_cols[x_col] + " + " + select_cols[y_col])
        subplots[count].scatter(stone_data[select_cols[x_col]], stone_data[select_cols[y_col]], alpha=0.5, color = 'red')
        subplots[count].scatter(soil_data[select_cols[x_col]], soil_data[select_cols[y_col]], alpha=0.5, color = 'blue')
        count+=1
plt.show()
```
